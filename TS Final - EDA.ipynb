```python
# Plot 'PM2.5 AQI Value' over time
plt.figure(figsize=(10, 6))
plt.plot(df1['Date'], df1['PM2.5 AQI Value'])
plt.title('PM2.5 AQI Value over Time')
plt.xlabel('Date')
plt.ylabel('PM2.5 AQI Value')
plt.show()
```

```python
# Plot 'Ozone AQI Value' over time
plt.figure(figsize=(10, 6))
plt.plot(df2['Date'], df2['Ozone AQI Value'])
plt.title('Ozone AQI Value over Time')
plt.xlabel('Date')
plt.ylabel('Ozone AQI Value')
plt.show()
```

```python
# Examine the frequency of each 'Site Name'
plt.figure(figsize=(10, 6))
sns.countplot(y=df1['Site Name'])
plt.title('Frequency of each Site Name - PM2.5')
plt.xlabel('Count')
plt.ylabel('Site Name')
plt.show()
```

```python
# Examine the frequency of each 'Site Name'
plt.figure(figsize=(10, 6))
sns.countplot(y=df2['Site Name'])
plt.title('Frequency of each Site Name - Ozone')
plt.xlabel('Count')
plt.ylabel('Site Name')
plt.show()
```

```python
# Examine the frequency of each 'Source'
plt.figure(figsize=(10, 6))
sns.countplot(y=df1['Source'])
plt.title('Frequency of each Source - PM2.5')
plt.xlabel('Count')
plt.ylabel('Source')
plt.show()
```

```python
# Examine the frequency of each 'Source'
plt.figure(figsize=(10, 6))
sns.countplot(y=df2['Source'])
plt.title('Frequency of each Source - Ozone')
plt.xlabel('Count')
plt.ylabel('Source')
plt.show()
```

```python
# PM2.5
# Assuming 'Date' is your column with dates
df1['Date'] = pd.to_datetime(df1['Date'])

# Set 'Date' as the index
df1.set_index('Date', inplace=True)

# Create a new DataFrame that includes all dates in the range
all_dates_df1 = pd.DataFrame(index=pd.date_range(start=df1.index.min(), end=df1.index.max()))

# Join the original DataFrame with the new one that includes all dates
df1 = all_dates_df1.join(df1)

# Forward fill the missing values
df1.ffill(inplace=True)
```

```python
#Ozonee
# Assuming 'Date' is your column with dates
df2['Date'] = pd.to_datetime(df2['Date'])

# Set 'Date' as the index
df2.set_index('Date', inplace=True)

# Create a new DataFrame that includes all dates in the range
all_dates_df2 = pd.DataFrame(index=pd.date_range(start=df2.index.min(), end=df2.index.max()))

# Join the original DataFrame with the new one that includes all dates
df2 = all_dates_df2.join(df2)

# Forward fill the missing values
df2.ffill(inplace=True)
```

```python
df1 = df1.drop(columns='Main Pollutant')
```
